```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Dropout, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Nadam
from tensorflow.keras.utils import plot_model
from tensorflow.keras.regularizers import l2  # Import L2 regularization
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import RandomOverSampler
from sklearn.preprocessing import StandardScaler
import optuna

# read dataset
(train_images, train_labels), (
    test_images,
    test_labels,
) = keras.datasets.mnist.load_data()
```

```python
train_images = train_images.astype("float32") / 255
test_images = test_images.astype("float32") / 255

train_images = np.expand_dims(train_images, axis=-1)
test_images = np.expand_dims(test_images, axis=-1)

train_labels = keras.utils.to_categorical(train_labels)
test_labels = keras.utils.to_categorical(test_labels)
```

```python
model = keras.Sequential(
    [
        keras.layers.Conv2D(
            32,
            kernel_size=(3, 3),
            activation="relu",
            input_shape=(28, 28, 1),
        ),
        keras.layers.MaxPooling2D(pool_size=(4, 4)),
        keras.layers.Flatten(),
        keras.layers.Dense(64, activation="relu"),
        keras.layers.Dense(32, activation="relu"),
        keras.layers.Dense(10, activation="softmax"),
    ]
)
```

```python
# Hier wordt gradient descent toegepast.
# De optimizer update de model weights tijdens het trainen. Het gebruikt gradient informatie van de loss.

model.compile(optimizer=SGD(learning_rate=0.15), loss='categorical_crossentropy', metrics=['accuracy', 'AUC', 'MeanSquaredError'])
```

```python
# Hier worden gradient descent en backpropagation uitgevoerd

history = model.fit(
    train_images,
    train_labels,
    epochs=50,
    batch_size=32,
    validation_data=(test_images, test_labels),
)
```

```python
test_loss, test_accuracy, test_auc, test_mean_squared_error = model.evaluate(test_images, test_labels)

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot Training and Test Loss
axes[0, 0].plot(history.history["loss"], label='Training Loss', color='b')
axes[0, 0].plot(history.history["val_loss"], label='Test Loss', color='r')
axes[0, 0].set_title('Loss')
axes[0, 0].set_xlabel('Epochs')
axes[0, 0].set_ylabel('Loss')
axes[0, 0].legend(loc='upper right')

# Plot Training and Test AUC
axes[0, 1].plot(history.history["auc"], label='Training AUC', color='b')
axes[0, 1].plot(history.history["val_auc"], label='Test AUC', color='r')
axes[0, 1].set_title('AUC')
axes[0, 1].set_xlabel('Epochs')
axes[0, 1].set_ylabel('AUC')
axes[0, 1].legend(loc='lower right')

# Plot Training and Test Mean Squared Error (MSE)
axes[1, 0].plot(history.history["mean_squared_error"], label='Training MSE', color='b')
axes[1, 0].plot([test_mean_squared_error] * len(history.history["mean_squared_error"]), label='Test MSE', color='r')
axes[1, 0].set_title('Mean Squared Error (MSE)')
axes[1, 0].set_xlabel('Epochs')
axes[1, 0].set_ylabel('MSE')
axes[1, 0].legend(loc='upper right')

# Plot Training and Test Accuracy
axes[1, 1].plot(history.history["accuracy"], label='Training Accuracy', color='b')
axes[1, 1].plot([test_accuracy] * len(history.history["accuracy"]), label='Test Accuracy', color='r')
axes[1, 1].set_title('Accuracy')
axes[1, 1].set_xlabel('Epochs')
axes[1, 1].set_ylabel('Accuracy')
axes[1, 1].legend(loc='upper right')

plt.tight_layout()
plt.show()
print("Test Loss:", test_loss)
```

**Why do they shuffle the dataset?**

Het shuffelen van een dataset wordt vaak gedaan voor de training, omdat sommige datasets gesorteerd staan op een feature. Door te shuffelen, krijg je een volledig willekeurige volgorde. Het wordt dus gedaan om te voorkomen dat het model de volgorde of patronen specifiek voor het dataset leert. Het shuffelen zorgt ervoor dat elke batch gegevens die aan het model wordt gevoed, een willekeurige selectie is uit je hele dataset. Dit helpt het model beter te generaliseren en overfitting te voorkomen. Als de gegevens niet werden geshuffelt, zou het model kunnen leren om goed te presteren op een specifieke volgorde van voorbeelden, wat niet zou generaliseren naar nieuwe gegevens.


**Why do they use a batch size of 32?**

Kleinere batch sizes kunnen meer willekeurigheid brengen bij het updaten van de weights, en kunnen het model sneller laten trainen in sommige gevallen. Ook hebben kleinere batch sizes minder geheugen, wat belangrijk kan zijn als je traint op een kleinere GPU of CPU. De beste batchgrootte kan verschillen afhankelijk van je dataset en je model. 32 is waarschijnlijk hun beste grootte gebleken, na het expirimenteren.

**What does 'tf.keras.optimizers.Adam(0.001)' mean?**

Dit selecteert de optimizer die wordt gebruikt voor training. In dit geval wordt Adam als optimizer gebruikt met een learning rate van 0,001. Adam is een populair algoritme dat de learning rate aanpast tijdens training op basis van eerdere gradients. De learning rate 0,001 is een hyperparameter die de grootte bepaalt tijdens de gradient descent. Het is een kleine waarde, wat betekent dat het model kleine stappen zal nemen in de richting van de (local) minimum cost.


```python
# read dataset
(train_images, train_labels), (
    test_images,
    test_labels,
) = keras.datasets.mnist.load_data()

train_images = train_images.astype("float32") / 255
test_images = test_images.astype("float32") / 255

train_images = np.expand_dims(train_images, axis=-1)
test_images = np.expand_dims(test_images, axis=-1)

train_labels = keras.utils.to_categorical(train_labels)
test_labels = keras.utils.to_categorical(test_labels)
```

```python
    
    
def create_model(trial):
    model = Sequential()
    model.add(Conv2D(32, kernel_size=(3, 3), activation="relu", input_shape=(28, 28, 1)))
    model.add(MaxPooling2D(pool_size=(4, 4)))
    model.add(Flatten())
    model.add(Dense(64, activation="relu"))
    model.add(Dense(32, activation="relu"))
    model.add(Dense(10, activation="softmax"))

    lr = trial.suggest_float("lr", 1e-3, 1e-1, log=True)

    optimizer_name = trial.suggest_categorical("optimizer", ["SGD", "Adam", "RMSprop"])
    
    num_epochs = trial.suggest_int("num_epochs", 10, 20)
    
    if optimizer_name == "SGD":
        optimizer = SGD(learning_rate=lr)
    elif optimizer_name == "Adam":
        optimizer = Adam(learning_rate=lr)
    elif optimizer_name == "RMSprop":
        optimizer = RMSprop(learning_rate=lr)

    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

    return model, num_epochs
```

```python
def objective(trial):
    # Create and compile the model
    model, num_epochs = create_model(trial)

    # Train the model
    history = model.fit(
        train_images,
        train_labels,
        epochs=num_epochs,
        batch_size=256,
        validation_data=(test_images, test_labels),
        verbose=0,  # Set to 1 for training details
    )

    # Evaluate the model
    _, test_accuracy = model.evaluate(test_images, test_labels)

    return test_accuracy
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=10)  # You can adjust the number of trials
```

```python
best_trial = study.best_trial
print("Best Trial:")
print("  Value: {}".format(best_trial.value))
print("  Params: ")
for key, value in best_trial.params.items():
    print("    {}: {}".format(key, value))

# Train the final model with the best hyperparameters
best_model, best_epochs = create_model(best_trial)
history = best_model.fit(
    train_images,
    train_labels,
    epochs=best_epochs,
    batch_size=32,
    validation_data=(test_images, test_labels),
)
```

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Model Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Train", "Test"], loc="upper left")
plt.show()

train_loss = history.history['loss']
val_loss = history.history['val_loss']

iterations = range(1, len(train_loss) + 1)

plt.figure(figsize=(10, 6))
plt.plot(iterations, train_loss, label='Training Loss')
plt.plot(iterations, val_loss, label='Validation Loss')
plt.title('Loss vs. Iterations')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
